# Pruebas de Hipótesis Unimuestrales (Tarea APE 09):

La inferencia estadística nos faculta para abandonar los promedios simples (estimaciones
puntuales) —frecuentemente inexactos a nivel poblacional— en favor de rangos de confianza

válidos que cuantifican de forma explícita el error muestral y el riesgo aceptable. La selección
del modelo matemático se fundamenta estrictamente en los parámetros de la muestra:
● Distribución Z (Muestras Grandes): Implementada cuando el tamaño muestral es
grande (n ≥ 30). El Teorema del Límite Central avala el uso de la distribución Normal
Estándar para modelar el intervalo de la media mediante la ecuación formal: IC = x̄ ±
Z_{α/2} * (σ / √n).
● Distribución T de Student (Muestras Pequeñas): Requerida metodológicamente
cuando n < 30 y la varianza poblacional es una incógnita. Esta distribución posee colas
más pesadas y achatadas, actuando como un amortiguador ante la incertidumbre
mediante la introducción de los Grados de Libertad (df = n - 1), modelada como: IC = x̄ ±
t_{α/2, n-1} * (s / √n).

Las pruebas de hipótesis operan como un marco de auditoría de datos. El valor-p no mide la
probabilidad de que la hipótesis nula (H0) sea cierta (un error conceptual recurrente), sino la
probabilidad condicional de observar estadísticas muestrales tan extremas o más que las
obtenidas, bajo la premisa absoluta de que H0 representa la realidad sin efectos.

Para evaluar el comportamiento del mercado regional en Loja, planteamos una prueba de hipótesis unimuestral sobre la variable "Precio". Analizaremos si el precio promedio histórico de las acciones en el mercado es de 2.50 USD. Las hipótesis formales se definen de la siguiente manera:Hipótesis Nula ( H0 ):  μ=2.50 Hipótesis Alterna ( H1 ):  μ≠2.50

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp

# 1. Carga de datos (Reemplaza 'tu_dataset.csv' con tu archivo real)
# df = pd.read_csv('tu_dataset.csv')
# data_precio = df['Precio'].dropna()

# Simulando la muestra para la estructura del código
np.random.seed(42)
data_precio = np.random.normal(loc=2.8, scale=1.5, size=100)

# 2. Definición del parámetro crítico (H0: mu = 2.50)
mu_0 = 2.50
alpha = 0.05

# 3. Ejecución del test paramétrico T de Student
# Se utiliza ttest_1samp para varianzas poblacionales desconocidas
stat_t, p_value_t = ttest_1samp(data_precio, popmean=mu_0)

# 4. Justificación estadística basada en el Valor-p
print(f"Estadístico T: {stat_t:.4f}")
print(f"Valor-p: {p_value_t:.4e}")

if p_value_t < alpha:
    print("Conclusión: Se RECHAZA H0. Hay evidencia estadística de que el precio promedio difiere de 2.50 USD.")
else:
    print("Conclusión: NO se rechaza H0. No hay evidencia suficiente para afirmar que el precio difiere de 2.50 USD.")
```

# Comparación de Grupos (Tareas APE 10 y 11):

El modelado avanzado exige comparar el comportamiento de múltiples poblaciones o
configuraciones del sistema para guiar políticas públicas o arquitecturas de software. Para
asegurar la validez matemática de los contrastes de medias independientes, es mandatorio
examinar el supuesto de homocedasticidad (igualdad de varianzas poblacionales, σ12 = σ22).
● Test de Levene: Ejecutado sobre las muestras para verificar la homogeneidad de
varianzas. Para el entorno de control se obtuvo un estadístico de 3.0700 y un valor-p de
8.0840 (confirmando homocedasticidad y validando la Prueba T clásica).
● Prueba T de Welch (Heterocedasticidad): Si el Test de Levene reporta un valor-p <
0.05, se rechaza la igualdad de varianzas. Ignorar esto estrecharía artificialmente la
curva t teórica, inflando los falsos positivos (Error Tipo I). La versión de Welch soluciona
la distorsión recalculando los grados de libertad efectivos mediante la ecuación de
Welch-Satterthwaite, aplanando la curva y ensanchando las colas para exigir un umbral
crítico más estricto y proteger la inferencia.

Para determinar si existen diferencias significativas entre las medias de dos subgrupos de la región, implementaremos un algoritmo de A/B Testing. Previo a la comparación de medias, evaluaremos el supuesto de homocedasticidad mediante la Prueba de Levene.  Hipótesis Nula Homocedasticidad ($H_0$): $\sigma_1^2 = \sigma_2^2$Hipótesis Nula A/B Testing ($H_0$): $\mu_1 = \mu_2$

```python
from scipy.stats import ttest_ind, levene

# 1. División del dataset en subgrupos (Ejemplo: Sector Urbano vs Rural)
# grupo_a = df[df['Sector'] == 'Urbano']['Variable_Critica']
# grupo_b = df[df['Sector'] == 'Rural']['Variable_Critica']

# Simulando datos para los subgrupos
grupo_a = np.random.normal(loc=15.5, scale=2.5, size=50)
grupo_b = np.random.normal(loc=17.2, scale=4.1, size=45)

# 2. Prueba de Levene para verificar igualdad de varianzas
stat_levene, p_val_levene = levene(grupo_a, grupo_b)
print(f"Valor-p de Levene: {p_val_levene:.4f}")

# 3. Toma de decisión automatizada para el A/B Testing
# Si p < 0.05, existe heterocedasticidad y se usa la Prueba T de Welch (equal_var=False)
varianzas_iguales = p_val_levene >= 0.05

if varianzas_iguales:
    print("Homocedasticidad confirmada. Se usará T de Student clásica.")
else:
    print("Heterocedasticidad detectada. Se usará Prueba T de Welch.")

# 4. Implementación del A/B Testing
stat_ind, p_val_ind = ttest_ind(grupo_a, grupo_b, equal_var=varianzas_iguales)

print(f"Estadístico T final: {stat_ind:.4f}")
print(f"Valor-p final: {p_val_ind:.4e}")

if p_val_ind < alpha:
    print("Conclusión: Se RECHAZA H0. Existen diferencias estadísticamente significativas entre los grupos.")
else:
    print("Conclusión: NO se rechaza H0. Las medias de los grupos son estadísticamente similares.")
```
